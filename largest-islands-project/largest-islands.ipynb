### Project Ideas

- What are the 10 largest islands in the tropics?

- What are the largest islands in each `region`?

- Create a line graph with `area` on the y-axis and `rank` on the x-axis. The data should be ordered by `rank`, from largest to smallest.

- What islands are composed of multiple countries?
	- Hint: try `countries.str.contains(",")`

```python
# Top 10 largest islands in the tropics are:
import pandas as pd

df = pd.read_csv('largest-islands.csv')
df = df.query("climate == 'tropics'")
df.sort_values(by = 'area', ascending = False).head(10)
```

```python
# Largest island in each region include:
import pandas as pd

df = pd.read_csv('largest-islands.csv')
df.loc[df.groupby('region')['area'].idxmax()]

```

```python
# A line graph mapping the area to the rank from largest to smallest
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('largest-islands.csv')
df_sorted = df.sort_values(by='rank')
df_sorted.plot(x='rank', y='area', kind='line', title='Island Area by Rank', xlabel='rank', ylabel='area')
plt.show()
```

```python
# islands that are composed of multiple countries include:

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('largest-islands.csv')
df = df[df['countries'].str.contains(",")]
df
```
